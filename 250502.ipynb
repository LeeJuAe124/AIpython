```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 예시 데이터: 월별 기온 (꺾은 선 그래프)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
# 딕셔너리 표현도 가능
temperatures = [5, 7, 12, 18, 22]

plt.plot(months, temperatures)          # x축 데이터, y축 데이터 순서로 입력
plt.title('Monthly Temperature Change') # 그래프 제목
plt.xlabel('Month')                     # x축 이름
plt.ylabel('Temperature (C)')           # y축 이름
plt.show()                              # 그래프 보여주기
```

```python
# 예시 데이터: 키와 몸무게 (산점도)
height = [160, 165, 170, 175, 180]
weight = [55, 60, 68, 72, 78]

plt.scatter(height, weight) # x축 데이터, y축 데이터
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
# 예시 데이터: 과일별 판매량 (막대 그래프)
fruits = ['Apple', 'Banana', 'Orange']
sales = [100, 150, 80]

plt.bar(fruits, sales) # x축 카테고리, y축 값
plt.title('Fruit Sales')
plt.xlabel('Fruit')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
# 예시 데이터: 학생들의 시험 점수 (히스토그램)
scores = [65, 72, 75, 78, 80, 82, 85, 88, 90, 91, 93, 95, 98, 100]

plt.hist(scores, bins=5, edgecolor='black') # 데이터, 구간(bin) 개수 설정, edgecolor로 테두리획 추가
plt.title('Exam Score Distribution')
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.show()
```

```python
import matplotlib.pyplot as plt

year = [2000, 2005, 2010, 2015, 2020]
single_person = [15.5, 20.0, 23.9, 27.2, 31.7]
baby_born = [640089, 438707, 470171, 438420, 272337]

plt.figure(figsize=(5,3))
 # 1인 가구 비율(single_person)과 출생아 수(baby_born) 사이의 관계를 나타내는 산점도를 그립니다.
for i in range(len(year)):
  plt.annotate(year[i],xy=(single_person[i], baby_born[i]),xytext=(5,-10),textcoords='offset points') #각 연도에 해당하는 데이터 지점 옆에 해당 연도를 표시하는 주석을 추가
plt.xlim(14,35)
plt.ylim(230000,700000)
plt.xlabel('single-person household ratio')
plt.ylabel('the number of babies born')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sports = ['Soccer', 'Basketball', 'Baseball', 'Running']
students = [8, 12, 5, 10]
data = pd.DataFrame({'Sport': sports, 'Students': students})

sns.set(style="whitegrid")

plt.figure(figsize=(8, 6))
sns.barplot(x='Sport', y='Students', data=data, palette='pastel')

plt.title('학생들이 선호하는 운동', fontsize=16, fontweight='bold')
plt.xlabel('운동', fontsize=14)
plt.ylabel('학생 수', fontsize=14)

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
