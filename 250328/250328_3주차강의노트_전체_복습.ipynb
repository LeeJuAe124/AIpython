```python
import numpy as np

# 배열 생성
arr = np.array([[1, 2, 3, 4, 5],])
print(arr)

# 배열 연산
arr2 = arr * 2
print(arr2)

# 통계 함수
mean_value = np.mean(arr)
print(mean_value)
```

```python
import pandas as pd

# DataFrame 생성
data = {'name': ['Alice', 'Bob', 'Charlie'],
        'age': [25, 30, 35],
        'city': ['New York', 'Los Angeles', 'Paris']}
df = pd.DataFrame(data)
print(df)

# CSV 파일 읽기
# df = pd.read_csv('data.csv')

# 데이터 선택
print(df['name'])  # 'name' 열 선택

# 데이터 필터링
df_filtered = df[df['age'] > 28]
print(df_filtered)  # 'age'가 28보다 큰 행 선택
```

```python
# 필요한 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 샘플 CSV 파일 생성(실제 환경에서는 외부 파일 로드)
# 존재하는 CSV 파일이 있다면 이 코드는 생략하고 pd.read_csv()로 불러오기 가능
sample_data = {
    'category': ['A', 'B', 'C', 'D', 'E'],
    'Value': [45, 78, 32, 95, 63]
}
df = pd.DataFrame(sample_data)
df.to_csv('sample_data.csv', index=False)
print("샘플 CSV 파일 생성 완료")

# CSV 파일 읽기
data = pd.read_csv('sample_data.csv')
print("\n데이터 미리보기: ")
print(data.head())

# 기본 통계 계산
mean_value = data['Value'].mean()
sum_value = data['Value'].sum()
max_value = data['Value'].max()
min_value = data['Value'].min()

# 통계 결과 출력
print("\n기초 통계 분석 결과: ")
print(f"평균값: {mean_value: .2f}")
print(f"합계: {sum_value}")
print(f"최대값: {max_value}")
print(f"최소값: {min_value}")
print("\n상세 통계 요약: ")
print(data.describe())

# 데이터 시각화: 막대 그래프
plt.figure(figsize=(10, 6))
bars = plt.bar(data['category'], data['Value'], color='skyblue')

# 그래프 제목 및 레이블 설정
plt.title('카테고리별 값 분포', fontsize=16)
plt.xlabel('카테고리', fontsize=12)
plt.ylabel('값', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 각 막대에 값 표시
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height+1,
            f'{height}', ha='center', va='bottom')

# 그래프 저장 및 표시
plt.tight_layout()
plt.savefig('category_values_chart.png')
plt.show()

print("\n분석 그래프가 'category_values_chart.png'로 저장되었습니다.")
```
