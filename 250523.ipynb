```python
import pandas as pd
from sklearn.datasets import load_iris

# Iris 데이터셋 로드
iris = load_iris()

# 데이터프레임으로 변환
iris_data = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# 컬럼 이름을 간단하게 바꾸기
iris_data.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

print(iris_data.sort_values(by=['petal_length'], ascending=False))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=iris_data['petal_length'], y=iris_data['petal_width'])
plt.show()
```

```python
data = iris_data['petal_length']

# IQR 계산
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

# 이상치 범위 정의
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 이상치 찾기
outliers = data[(data < lower_bound) | (data > upper_bound)]

# 결과 출력
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("하한:", lower_bound)
print("상한:", upper_bound)
print("이상치:", outliers.tolist())

import matplotlib.pyplot as plt
plt.scatter(x=iris_data['petal_length'], y=iris_data['petal_width'])
plt.show()
```

```python
import pandas as pd

# 데이터
data = pd.Series([10, 11, 12, 13, 14, 15, 100, 101])

# Q1, Q3 계산
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)

# IQR 계산
IQR = Q3 - Q1

# 하한, 상한 계산
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"하한: {lower_bound}")
print(f"상한: {upper_bound}")

# 이상치 찾기
outliers = data[(data < lower_bound) | (data > upper_bound)]
print("이상치:\n", outliers)
```
