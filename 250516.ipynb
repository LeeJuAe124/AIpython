```python
# 문제 5: 이상치 확인하기
# 데이터프레임에서 이상치가 있는지 확인하세요.
outliers = df[(df['점수'] > 100) | (df['공부시간'] > 10)]
print(outliers)
```

```python
# 문제 6: 중복 데이터 확인하기
# 데이터프레임에서 중복된 행이 있는지 확인하세요
print("중복 제거 전 데이터프레임 크기:", df.shape)

num_duplicates = df.duplicated(keep=False).sum()
print("완전 중복 행 개수:", num_duplicates)

duplicates = df[df.duplicated(keep=False)]
print("완전 중복 행:\n", duplicates)
```

```python
# 문제 7: 중복 데이터 삭제하기
# 데이터프레임에서 중복된 데이터를 삭제하고, 중복이 제거된 데이터프레임을 출력하세요
df_deduplicated = df.drop_duplicates()
print(df_deduplicated)
```

```python
# 문제 9: 시각화
# 공부시간과 점수에 대한 산점도 그래프를 작성하시오
import matplotlib.pyplot as plt
import matplotlib as mpl

plt.rcParams['font.family'] = 'AppleGothic'
mpl.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(12, 8))
plt.scatter(df_deduplicated['공부시간'], df_deduplicated['점수'], color='blue', alpha=0.7)
plt.title('원본 데이터: 공부시간 vs 점수', fontsize=14)
plt.xlabel('공부시간', fontsize=12)
plt.ylabel('점수', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# 점수의 이상치 제거
Q1_score = df['점수'].quantile(0.25)
Q3_score = df['점수'].quantile(0.75)
IQR_score = Q3_score - Q1_score
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

# 공부시간의 이상치 제거
Q1_time = df['공부시간'].quantile(0.25)
Q3_time = df['공부시간'].quantile(0.75)
IQR_time = Q3_time - Q1_time
lower_bound_time = Q1_time - 1.5 * IQR_time
upper_bound_time = Q3_time + 1.5 * IQR_time

outliers = df[(df['점수'] < lower_bound) | (df['점수'] > upper_bound)]
print("점수 이상치:\n", outliers)

print()

outliers = df[(df['공부시간'] < lower_bound_time) | (df['공부시간'] > upper_bound_time)]
print("공부시간 이상치:\n", outliers)

print()

# 이상치 제거
df_cleaned = df[
    (df['점수'] >= lower_bound_score) & (df['점수'] <= upper_bound_score) &
    (df['공부시간'] >= lower_bound_time) & (df['공부시간'] <= upper_bound_time)
]

print("이상치 제거 후 데이터프레임:")
print(df_cleaned)
```
